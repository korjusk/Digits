### Goal
The goal in this competition is to take an image of a handwritten single digit, and determine what that digit is.
For every ImageId in the test set, you should predict the correct label.

#### Baseline prediction with 10% accuracy: Every number is '9'

```python
import os

for i in range(10):
    p = f'data/train/{str(i)}'
    if not os.path.exists(p):
        os.makedirs(p)
    
    p = f'data/valid/{str(i)}'
    if not os.path.exists(p):
        os.makedirs(p)
```

```python
p = 'data/test1'
if not os.path.exists(p):
    os.makedirs(p)
```

```python
for i in range(len(df)):
    digit = df.iloc[i]
    label = digit.label
    digit = digit.drop(['label'])
    digit = digit.values.tolist()
    img = Image.new('L', (28, 28))
    img.putdata(digit)
    
    if (i % 10):
        p = f'data/train/{label}/{label}.{i}.png'
    else:
        p = f'data/valid/{label}/{label}.{i}.png'
        
    img.save(p)
```

```python
for i in range(len(test_df)):
    digit = test_df.iloc[i]
    digit = digit.values.tolist()
    img = Image.new('L', (28, 28))
    img.putdata(digit)
    p = f'data/test1/{i}.png'
    img.save(p)
```
